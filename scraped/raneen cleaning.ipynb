```python
import pandas as pd 
import numpy as np
import re
```

```python
df = pd.read_csv('raneen_products.csv')
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
def extract_first_value(text):
    return re.findall(r'\d{1,3}(?:,\d{3})*', text)[0]
```

```python
df['current_price'] = df['current_price'].apply(extract_first_value)
```

```python
df['current_price'] = df['current_price'].str.replace('جنيه', '', regex=False).str.replace(',', '').astype(float)
df['old_price'] = df['old_price'].str.replace('جنيه', '', regex=False).str.replace(',', '').astype(float)
```

```python
df['discount'] = ((df['old_price'] - df['current_price']) / df['old_price']) * 100
```

```python
def format_row(row):
    return (
        f"المنتج هو '{row['name']}'."
        f"السعر هو {row['current_price']}. "
        f"الخصم المتاح هو {row['discount']}. "
    )
```

```python
df['Formatted'] = df.apply(format_row, axis=1)
```

```python
df[['Formatted', 'link']].to_csv('Raneen.csv', index=False , encoding='utf-8-sig')
```
