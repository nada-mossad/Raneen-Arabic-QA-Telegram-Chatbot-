```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
)
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
```

```python
urls = [
    "https://www.raneen.com/ar/appliances/large-appliances/refrigerators-and-freezers/refrigerators",
    "https://www.raneen.com/ar/appliances/large-appliances/refrigerators-and-freezers/freezers",
    "https://www.raneen.com/ar/appliances/large-appliances/cookers",
    "https://www.raneen.com/ar/appliances/large-appliances/water-heaters",
    "https://www.raneen.com/ar/appliances/large-appliances/dish-washers" , 
    "https://www.raneen.com/ar/appliances/large-appliances/washers-and-dryers" ,
    "https://www.raneen.com/ar/furniture/living-room/tv-tables" , 
    "https://www.raneen.com/ar/furniture/living-room/sofas",
    "https://www.raneen.com/ar/furniture/bed-room/wardrobe" ,
    "https://www.raneen.com/ar/lighting-home-decor/home-decor" ,
    "https://www.raneen.com/ar/furniture/storage-units/shoe-rack" ,
    "https://www.raneen.com/ar/lighting-home-decor/lighting" , 
    "https://www.raneen.com/ar/furniture/bed-room" , 
    "https://www.raneen.com/ar/kitchen/food-dessert-serving-sets/dinner-dining-set",
    "https://www.raneen.com/ar/kitchen/food-keeper-jars",
    "https://www.raneen.com/ar/textile/comforters",
    "https://www.raneen.com/ar/fashion/women" , 
    "https://www.raneen.com/ar/fashion/for-kids",
    "https://www.raneen.com/ar/fashion/men",
    "https://www.raneen.com/ar/textile/quilt" ,
    "https://www.raneen.com/ar/textile/bed-sheets" , 
    "https://www.raneen.com/ar/textile/rugs"
    "https://www.raneen.com/ar/home/cleaning-tools-supplies"
    
]
```

```python
data = []

for url in urls:
    driver.get(url)
    time.sleep(5)

    products = driver.find_elements(By.CSS_SELECTOR, ".product-item-info")

    for product in products:
        try:
            name_elem = product.find_element(By.CSS_SELECTOR, "a.product-item-link")
            name = name_elem.text.strip() or None
            link = name_elem.get_attribute("href") or None
        except:
            name = None
            link = None

        try:
            current_price = product.find_element(By.CSS_SELECTOR, "span.current-price").text.strip() or None
        except:
            current_price = None

        try:
            old_price = product.find_element(By.CSS_SELECTOR, "span.old-price").text.strip() or None
        except:
            old_price = current_price

        data.append({
            "name": name,
            "current_price": current_price,
            "old_price": old_price,
            "link": link
        })

driver.quit()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("raneen_products.csv", index=False , encoding='utf-8-sig')
```
